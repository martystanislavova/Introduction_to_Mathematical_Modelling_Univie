Zeichnen  Sie  die  Ecken  des Polyeders $Ax \leq b,  \,  x \geq 0$ mit
		$$  A=  \begin{pmatrix} 
		1 & 1 & 1  \\
		1 & 0& 0  \\

		\end{pmatrix}, 
        \quad b =  \begin{pmatrix} 
		4  \\
		2  \\
		3  \\
		6  \\
		\end{pmatrix}. $$
		
und bestimmen  Sie  dessen Basen.
		
In welcher Ecke wird der Wert der Zielfunktion $- x_1 - 5 \, x_2 + 2 \,x_3 $  am größten?

### Lösung

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Define the coefficients of the constraints
A = np.array([[1, 1, 1], [1, 0, 0], [0, 0, 1], [0, 3, 1]])
b = np.array([4, 2, 3, 6])

# Define the feasible region
x1 = np.linspace(0, 5, 100)
x2 = np.linspace(0, 5, 100)
x1, x2 = np.meshgrid(x1, x2)
x3 = (b[0] - A[0,0]*x1 - A[0,1]*x2) / A[0,2]
    
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
    
# Plot the feasible region
ax.plot_surface(x1, x2, x3, alpha=0.5)

# Plot the vertices of the polyhedron
vertices = np.array([[0,0,0], [2,0,3], [3,0,0], [0,2,1], [0,0,6]])
ax.scatter(vertices[:,0], vertices[:,1], vertices[:,2], c='r')

# Set the axis labels
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# Set the viewing angle
ax.view_init(elev=25, azim=-135)

plt.show()
```
